# Watch Me Code 1: Seaborn

We will demonstrate Pythons data visualization library Seaborn


```python
# imports
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/mafudge/datasets/master/exam-scores/exam-scores.csv")
df.sample(10)
```

## Barplots

- What are the average grade by section?
- Exam version?
- Both?

```python
sns.barplot(data=df, y="Exam_Version", x="Student_Score", errorbar=None, estimator="mean")
```

```python
sns.barplot(data=df, x="Class_Section", y="Student_Score", hue="Class_Section", estimator="mean", errorbar=None)
```

```python
sns.barplot(data=df, x="Exam_Version", y="Student_Score", hue="Exam_Version", estimator="mean", errorbar=None)
```

```python
sns.barplot(data=df, x="Class_Section", y="Student_Score", hue="Exam_Version", estimator="mean", errorbar=None)
```

```python
## Counts: How manuy students got each exam?
sns.barplot(data=df, x="Class_Section", y="Student_Score", hue="Exam_Version", estimator="count", errorbar=None)
```

```python
# better scores on avergae if you studied in groups?
sns.barplot(data=df, x="Studied_In_Groups", y="Student_Score")
```

## Line Plots

There's no time series here but we do have implicit ordering with completion time so we can place that on the X-Axis.

```python
df.sample(5)
```

```python
sns.lineplot(data=df, x="Completion_Time", y="Student_Score")
```

## Scatter plot

Maybe this would look better as a scatter? No visible correlation. The line should slope up or down.

```python
sns.lmplot(data=df, x="Completion_Time", y="Student_Score")
```

```python
# lets add in a hue
sns.lmplot(data=df, x="Completion_Time", y="Student_Score", hue="Studied_In_Groups")
```

```python
# lets add in a hue
sns.lmplot(data=df, x="Completion_Time", y="Student_Score", hue="Did_Exam_Prep Assignment")
```

```python
# lets add in a hue
sns.lmplot(data=df, x="Completion_Time", y="Student_Score", hue="Made_Own_Study_Guide")
```

```python
filtered = df[ df['Made_Own_Study_Guide'] != '?' ]
sns.histplot(data=filtered, x="Student_Score", hue="Made_Own_Study_Guide")
```
