## Visualization Heuristics

- A **dimension** goes on the X-Axis. This is some category or time-oriented value such as product type or order date.
- A **measure** goes on the Y-Axis such as order amount or product quantity
- Another dimension can be added to the **legend** where we plot the same measure across another group of values. Example: total orders by customer and product type.


## Kinds of Visualizations

- **barplot** - categorical data where there is not relationship between the categories.
- **lineplot** - for time oriented values or other values where there is a seqential relationship.
- **scatterplot** - for comparison of two measures
- **histogram** - for counting the frequency of values, placing them into buckets or ranges.
