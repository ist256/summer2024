```python
headers = {"User-Agent": "Mozilla/5.0"}
tables = pd.read_html("https://www.baseball-reference.com/leagues/MLB-standings.shtml", storage_options=headers)
len(tables)
```

```python
from IPython.display import display
for index in range(len(tables)):
    print("INDEX:", index)
    display(tables[index])
```

```python
tables[0]['League'] = 'AL'
tables[0]['Division'] = 'East'
```

```python
teams_df = pd.concat(tables, ignore_index=True)
```

```python
teams_df
```

```python
# think of loc as in location not index. This says row 25 column label TM
teams_df.loc[25, "Tm"] # for more on loc...www.youtube.com/watch?v=lJDtzZsmF0g
```

```python
# Everone who has more than 5 wins
teams_df[ teams_df['W'] > 5 ]
```
